```python
# libraries
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Create a dataset
df = pd.read_csv('s2_table.csv')
df = df.set_index(df.columns[0])
df=df.T
# plot a heatmap with annotation
ax=sns.heatmap(df, annot=True, annot_kws={"size": 7})
ax.invert_yaxis()
```
