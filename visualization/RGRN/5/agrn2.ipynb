```python
from pyvis import network as net
from IPython.core.display import display, HTML
import pandas as pd
df= pd.read_csv("./final_go_5.csv", names=["source", "target", "FI"])
gold1 = pd.read_csv("./gold5.csv", names=["source", "target", "FI"])

# threshold= 0.1
# df=df.loc[df['FI'] >= threshold]


df.sort_values(by=['FI'], inplace=True, ascending=False)
df = df.head(100)

gold1.sort_values(by=['FI'], inplace=True, ascending=False)
gold1=gold1.loc[gold1['FI'] == 1]

df = df.drop(labels='FI', axis=1)
gold1 = gold1.drop(labels='FI', axis=1)
# gold1 = gold1.head(100)

gold1.to_csv('gold.csv', sep='\t', index=False)
```

```python
columns=["source", "target"]
# print(df)
all_values = []
for column in columns:
    this_column_values = df[column].tolist()
    all_values += this_column_values

one_column_df = pd.DataFrame(all_values)
# one_column_df = one_column_df.iloc[1:]
one_column_df = one_column_df.drop(labels=0, axis=0)
# print(one_column_df)
```

```python
all_values_gold = []
for column in columns:
    this_column_values = gold1[column].tolist()
    all_values_gold += this_column_values

one_column_gold = pd.DataFrame(all_values_gold)
one_column_gold = one_column_gold.iloc[0:]
```

```python
g=net.Network(directed=True,height='4000px', width='90%',heading='')
# nodes =node[0]

nodes =one_column_df[0]
# print(nodes)
# g.add_nodes(nodes) # node ids and labels = ["a", "b", "c", "d"]
g.add_nodes(nodes ) # node ids and labels = ["a", "b", "c", "d"]

df_concat = pd.concat([df, gold1])
df_concat = df_concat.reset_index(drop=True) 
df_group = df_concat.groupby(list(df_concat.columns)) 
idx = [x[0] for x in df_group.groups.values() if len(x) > 1]
df_blue =df_concat.reindex(idx)
df_blue.to_csv('blue1.csv', sep='\t', index=False)
# print(df_blue)
```

```python
r=0
b=0
for index, row in df.iterrows():
    if (row['source'], row['target']) in df_blue.values : 
        blue=(row['source'], row['target'])
        g.add_edge(row['source'], row['target'], color='blue')
        b=b+1
        print("blue",blue)
        

    else:
        red=(row['source'], row['target'])
        g.add_edge(row['source'], row['target'], color='red')
        r=r+1
        print("red",red)    

print("r",r) 
print("b",b)
```
