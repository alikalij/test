```python
import pandas as pd
from scipy import stats
# Read the csv file
data = pd.read_csv('GENIE3_5.csv')
```

```python
data
```

```python
data.sort_values(by=['target'])
```

```python
grouped_obj = data.groupby(["target"])
for key, item in grouped_obj:
    print("Key is: " + str(key))
    print(str(item), "\n\n")
    item.to_csv(str(key)+'.csv')
```

```python
z=data.groupby(["target"])["value"].std()
```

```python
m=data.groupby(["target"])["value"].mean()
```

```python
m.to_csv('mean.csv')
```

```python
for i in range (100):    
    data1 = pd.read_csv('G'+str(i+1)+'.csv')
    print(data1)
    zscores = stats.zscore(data1['value'])
    data1['z_score'] = zscores
    data1.to_csv('updated_G'+str(i+1)+'.csv')
```
